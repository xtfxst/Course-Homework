```
while True:
    acc = input('請輸入帳號?')
    pwd = input('請輸入密碼?')
    if (acc == 'abc' and pwd == '123'):
        print('帳號與密碼正確')
        break
    else:
        print('登入失敗')
```

```
password = "a123456"
times = 0
while times <= 3:
    answer = input("Enter Your Password: ") 
    if answer == password: 
        print("登入成功")
        break 
    else:
        times = times + 1
        chance = 3 - times
        print("密碼錯誤!還有", chance, "次機會")
```
